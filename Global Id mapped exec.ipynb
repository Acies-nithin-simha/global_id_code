```python
import pyodbc
import pandas as pd
import os
```

```python
# Define your DSN name
dsn_name = 'Nithin-GBQ-Connection'

# Create a connection using pyodbc
conn = pyodbc.connect(f'DSN={dsn_name};IgnoreTransactions=1')

# Define your query
# query = 'SELECT * FROM `gpc-dat-d-us-ba-digitala.DIGITALANALYTICS_BA_LAB.CUSTOMER_DATA`'

query = '''
  SELECT 
    unique_cust_id,
    STO_FIN_HRCHY__DCSTAR,
    month,
    SUM(sal__report_sale_amount) AS total_sale_amount,
    SUM(CASE WHEN sal__origination_code = '2' THEN sal__report_sale_amount ELSE 0 END) AS e_sale_amount
  FROM `gpc-dat-d-us-ba-digitala.DIGITALANALYTICS_BA_LAB.Global_id_mapper_q3_24`
  GROUP BY 
    unique_cust_id,
    STO_FIN_HRCHY__DCSTAR,
    month
'''


# Execute the query and read the data into a DataFrame
data = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the first few rows of the DataFrame
data.head(25)
```
